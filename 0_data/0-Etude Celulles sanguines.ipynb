**ETUDE DES CELLULES SANGUINES**

**Les cellules sanguines sont divisées en 3 catégories:**
- Les Globules Rouges appelés aussi hématies ou érythrocytes : 
     - ce sont les cellules les plus nombreuses du sang (5 millions/mm3)
     - Le rôle principal de ces cellules est d'assurer le transport de l'oxygène et du gaz carbonique entre les alvéoles pulmonaires et les tissus.


- Les Globules Blancs appelés aussi leucocytes : 
     - ils sont au nombre d'environ 8 000/mm3. 
     - ces cellules participent aux défenses spécifiques de l'organisme, aident à combattre les bactéries et les virus


- Les plaquettes appelés aussi thrombocytes  : 
     - Elles sont au nombre de 150 000 à 450 000/mm3
     - les plaquettes sont des fragments cytoplasmiques d'un mégacaryocyte
     - Elles jouent un rôle fondamental dans les phénomènes initiaux de coagulation
     

La numération globulaire (% de globules rouges, blancs et plaquettes) dans le sang permet de révéler des problèmes de santé : anémie, infection, hémorragie...

> La reconnaissance des différents types de globules blancs (leucocytes) va permettre d'identifier  la présence de certains types de leucocyte dans le sang qui n'y sont normalement pas présents, mais présents uniquement dans la moëlle osseuse.
>
> La présence de ces types de cellules dans le sang en quantité importante est le signe de problèmes de santé, tel que des leucémies

**Répartition des leucocytes (Globules blancs)**
- Les neutrophiles (55 à 70% des leucocytes)
- Les éosinophiles (1 à 4%)
- Les basophiles (0,5 à 1%)
- Les lymphocytes (20 à 40% des leucocytes) 
- Les monocytes (2 à 8 %)

**ETUDE DE LA LEUCEMIE**

> La leucémie est suspectée par des taux anormaux de cellules sanguines dans le sang : globules rouges, globules blancs polynucléaires (Basophil, Neutrophil, Eosinophil) et plaquettes <br>
>
> Le diagnostic est confirmé par un myeologramme : analyse des cellules de la moëlle osseuse 
>    
> Le diagnostic est confirmé si l’analyse montre que la moelle contient plus de 20 % de cellules immatures 
> L’analyse morphologique des cellules permet alors de définir la sous-catégorie de leucémie aiguë.
>
> Différentes leucémies : 
> - Leucémie aiguës
> - La leucémie lymphoïde chronique
> - La leucémie myéloïde chronique
>
> Stades de la maladie selon le nombre de cellules anormales dans le sang. Essentiellement lymphocytes anormaux?
